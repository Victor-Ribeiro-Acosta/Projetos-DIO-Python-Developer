```python
import pandas as pd
```

**Buscando base de dados**

```python
base_dados = pd.read_excel('AdventureWorks.xlsx')
base_dados.head()
```

**Processamento de Dados**

```python
base_dados.dtypes
```

*Nenhuma alteração precisa ser feita nos dados*

```python
base_dados.isnull().sum()
```

*Não há valres nulos na base de dados*

**Fazendo análises iniciais**

*Obtendo o custo total dos produtos vendidos*

```python
base_dados['Custo_total'] = base_dados['Custo Unitário'] * base_dados['Quantidade']
base_dados.head()
```

*Com o custo total, descobrimos o lucro líquido das vendas*

```python
base_dados['Lucro_liquido'] = base_dados['Valor Venda'] - base_dados['Custo_total']
base_dados.head()
```

*Extrair informações sobre o ano da venda e o tempo de envio dos produtos*

```python
base_dados['Tempo_envio'] = (base_dados['Data Envio'] - base_dados['Data Venda']).dt.days
base_dados['Ano_venda'] = base_dados['Data Venda'].dt.year
base_dados.head()
```

*Observando os tipos de dados gerados*

```python
base_dados['Tempo_envio'].describe()
```

*informações gerais sobre a receita*

```python
# Lucro médio
base_dados['Lucro_liquido'].mean()
```

```python
# Lucro Total
base_dados['Lucro_liquido'].sum()
```

**Visualizando As análises**

```python
import matplotlib.pyplot as plt
plt.style.use('seaborn')
```

*Vendas por loja*

```python
base_dados.groupby('ID Loja')['ID Produto'].count().sort_values(ascending = True).plot.barh()
plt.title('Total de vendas por loja')
plt.xlabel('Quantidade d vendas')
plt.ylabel('Loja')
```

*Total de vendas por marca*

```python
base_dados.groupby('Marca')['ID Produto'].count().plot.pie()
plt.title('Marcas de produtos')
plt.ylabel('')
```

*Total de Vendas por produto*

```python
base_dados.groupby('Produto')['ID Produto'].count().sort_values().plot.barh()
```

*Total de Vendas por cada classe*

```python
base_dados.groupby('Classe')['ID Produto'].count().plot.pie()
plt.title('Contagem de classes de produtos vendidos')
```

*Média de lucro por produto*

```python
base_dados.groupby('Produto')['Lucro_liquido'].mean().sort_values(ascending = False)
```

*Total de vendas por ano*

```python
base_dados.groupby('Ano_venda')['ID Produto'].count()
```

*Total de vendas por ano e por marca*

```python
base_dados.groupby(['Ano_venda', 'Marca'])['ID Produto'].count()
```

*Tempo de envio*

```python
plt.boxplot(base_dados['Tempo_envio']);
```
